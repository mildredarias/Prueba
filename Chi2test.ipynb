```c++
Double_t myLandau(Double_t* x, Double_t* par)
{
   if (par[2] <= 0) return 0; 
   Double_t den = ::ROOT::Math::landau_pdf( (x[0]-par[1])/par[2] ); 
   if (!par[3]) return den;
   return par[0]*den/par[2];
}
```
